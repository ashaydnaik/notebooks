A file is accessed through a Python file object:

```python
f = open("myfile", "w")
f.write("First line with necessary newline character\n")
```

The **open** statement creates a file object. <br>
Here, the file **myfile** in the current working directory is being opened in write ("w") mode. 

```python
f.write("Second line to write to the file\n")
f.close()
```

```python
f = open("myfile", "r")
```

After writing two lines to it and closing it, you open the same file again, this time in read ("r") mode.

```python
line1 = f.readline()
line2 = f.readline()
f.close()
print(line1, line2)
```
